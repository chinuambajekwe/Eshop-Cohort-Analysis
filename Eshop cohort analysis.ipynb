```python
#create a function to extract the month, year and first date of each month from invoice date 

def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
data["InvoiceDate"] = data["InvoiceDate"].apply(get_month)
```

```python
data
```

```python
#Create a new column, cohort date that indicates the first month
#each unique customer made their first purchase
def get_cohort_date(data):
    data["cohort_date"]= data.groupby("CustomerID")["InvoiceDate"].transform("min")
    return data["cohort_date"]

```

```python
data["cohort_date"] = get_cohort_date(data)
```

```python
data
```

### Cohort Index

```python
#create cohort index, represents the interval between first and next purchase 

def get_year_month(data, col):
    month = data[col].dt.month
    year = data[col].dt.year
    return month, year
```

```python
first_month, first_year = get_year_month(data, "cohort_date")
```

```python
first_month
```

```python
first_year
```

```python
latest_month, latest_year = get_year_month(data, "InvoiceDate")
```

```python
latest_month
```

```python
#Create Cohort Index 

#calculates the time difference in months btw the lastest and first purchase 
#for index we add multiply by 12 (to get the value in months)
# for customers who made purchase once, month_diff = 0, year difference = 0
# so we add 1 so index for customers who made a single purchase 
# so their index will be one

def create_cohort_index(first_month, first_year, latest_month, latest_year):
    year_diff = latest_year - first_year
    month_diff = latest_month - first_month
    index = year_diff * 12 + month_diff + 1 
    return index
```

```python
data["cohort_index"] = create_cohort_index(first_month, first_year, latest_month, latest_year)
```

```python
data
```

### Create a cohort table 


```python
cohort_info = data.groupby(["cohort_date", "cohort_index"])["CustomerID"].nunique().reset_index()
```

```python
cohort_info
```

```python
cohort_info.rename(columns = {"CustomerID" : "Number of Customers"}, inplace = True)
```

```python
cohort_info
```

### Create Pivot/Cohort table 

```python
cohort_table = cohort_info.pivot(index = "cohort_date", columns = ["cohort_index"], 
                                 values =  "Number of Customers")
                                
```

```python
cohort_table
cohort_table.index = cohort_table.index.strftime("%B %Y")
```

```python
cohort_table
```

```python
#heat map of table
plt.figure(figsize = (20,10))
sns.heatmap(cohort_table, annot = True, cmap = "viridis", fmt = ".2f" )
plt.show()
```

```python
#new cohort table 
new_cohort_table = cohort_table.divide(cohort_table.iloc[:, 0], axis = 0)
```

```python
new_cohort_table
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(new_cohort_table, annot = True, cmap = "viridis", fmt = ".0%" )

plt.savefig('plot5.jpg', format='jpg')
plt.show()
```
